```
df_orders_X_compradores_X_marketing_push = spark.sql("""

SELECT 
O.*,
P.event_channel,
P.event_name,
P.brand,
P.sample_type,
P.user_id,
P.external_user_id,
P.event_time_utc3,
P.platform,
P.campaign_id,
P.campaign_name,
P.message_variation_channel,
P.canvas_name,
P.canvas_step_id,
P.canvas_is_first_step,
P.canvas_first_step_name,
P.canvas_step_name,
P.canvas_real_step_name,
P.canvas_step_index,
P.canvas_tags,
P.send_id,
P.event_date,
P.dt,
P.DATA_ENVIO,
P.DATA_10MIN,
P.DATA_60MIN,
P.DATA_120MIN
FROM
TB_df_orders_X_compradores O
LEFT JOIN (
SELECT 
  *, 
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) AS DATA_ENVIO,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) + INTERVAL 10 MINUTE AS DATA_10MIN,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) + INTERVAL 60 MINUTE AS DATA_60MIN,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) + INTERVAL 120 MINUTE AS DATA_120MIN
FROM
TB_fia_marketing_push

) P ON P.external_user_id = O.customer_id


""")
```
