```
from pyspark.sql.functions import monotonically_increasing_id
#fia_orders x  fia_orders_with_cost_revenue
df1 = fia_orders.withColumn("order_number", monotonically_increasing_id())
df2 = fia_orders_with_cost_revenue.withColumn("order_number", monotonically_increasing_id())
DF_Full_orders_info = df2.join(df1, "order_number", "outer")

##fia_orders x  fia_orders_with_cost_revenue x df_fia_marketing_push
df1 = DF_Full_orders_info.withColumn("customer_id", monotonically_increasing_id())
df2 = df_fia_marketing_push.withColumn("customer_id", monotonically_increasing_id())
DF_Full_orders_info = df2.join(df1, "customer_id", "outer")

#fia_orders x  fia_orders_with_cost_revenue x df_fia_marketing_push x df_fia_customer_segmentation
df1 = DF_Full_orders_info.withColumn("customer_id", monotonically_increasing_id())
df2 = df_fia_customer_segmentation.withColumn("customer_id", monotonically_increasing_id())
DF_Full_orders_info = df2.join(df1, "customer_id", "outer")


#fia_orders x  fia_orders_with_cost_revenue x df_fia_marketing_push x df_fia_customer_segmentation x fia_sessions_visits
df1 = DF_Full_orders_info.withColumn("session_id", monotonically_increasing_id())
df2 = fia_sessions_visits.withColumn("session_id", monotonically_increasing_id())
DF_Full_orders_info = df2.join(df1, "session_id", "outer")



row_number = DF_Full_orders_info.count()
column_number = len(DF_Full_orders_info.dtypes)
print((row_number, column_number))

#fia_orders_with_cost_revenue = (492138, 8)
#fia_orders = (492138, 43)
#df_fia_marketing_push = (23113350, 22)
#df_fia_customer_segmentation = (210364, 50)
#fia_sessions_visits = (463620, 21)
#DF_Full_orders_info = (24001922, 141)
```

```
DF_Clicked_valido = spark.sql("""

SELECT 
  *, 
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) AS DATA_ENVIO_PUSH,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(order_timestamp_local,1,10),' ' ,SUBSTR(order_timestamp_local,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) AS DATA_order,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) + INTERVAL 10 MINUTE AS DATA_10MIN_PUSH,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) + INTERVAL 60 MINUTE AS DATA_60MIN_PUSH,
  CAST(UNIX_TIMESTAMP(CONCAT(SUBSTR(event_time_utc3,1,10),' ' ,SUBSTR(event_time_utc3,12,8)),'yyyy-MM-dd HH:mm:ss')  AS TIMESTAMP) + INTERVAL 120 MINUTE AS DATA_120MIN_PUSH

FROM
TD_DF_Clicked
where event_time_utc3 >= '2019-06-01T00:00:15.369Z'

""")

#display(DF_Clicked_valido)
```

```
df_push_X_order = spark.sql("""

SELECT 
DATA_order,DATA_ENVIO_PUSH, DATA_10MIN_PUSH,DATA_60MIN_PUSH,DATA_120MIN_PUSH,
CASE WHEN DATA_order BETWEEN DATA_ENVIO_PUSH AND DATA_10MIN_PUSH THEN 1 ELSE 0 END AS FL_COMPRA_10MIN,
CASE WHEN DATA_order BETWEEN DATA_ENVIO_PUSH AND DATA_60MIN_PUSH THEN 1 ELSE 0 END AS FL_COMPRA_60MIN,
CASE WHEN DATA_order BETWEEN DATA_ENVIO_PUSH AND DATA_120MIN_PUSH THEN 1 ELSE 0 END AS FL_COMPRA_120MIN

FROM
TD_DF_Clicked_valido
where order_timestamp_local != 'null'
""")
```
